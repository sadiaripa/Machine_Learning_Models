```python
url="https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/639388c2cbc2120a14dcf466e85730eb8be498bb/iris.csv"
dataset=pd.read_csv(url)
dataset.head(10)
```

```python
Species=pd.get_dummies(dataset["species"],drop_first=True)
Species.head(5)
```

```python
dataset=pd.concat([dataset,Species],axis=1)
dataset.drop("species",axis=1,inplace=True)
dataset.head(10)
```

```python
import random
import math
import pandas as pd
np.random.seed(42)
indices = np.random.permutation(len(dataset))
d=len(dataset)*0.67
d=math.ceil(d)
print(d)
n_training_samples = d
testing_data=len(dataset)-n_training_samples
count=0
learnset_data=pd.DataFrame()
learnset_target=pd.DataFrame()
while count<d:
    i=random.randint(0,len(dataset)-1)
    new_data=dataset.iloc[i,:4]
    learnset_data=learnset_data.append(new_data)
    new_target=dataset.iloc[i,4:]
    learnset_target=learnset_target.append(new_target)
    count+=1 
print("Number of learning data:",count)
print(learnset_data)
print(learnset_target)
test_data=pd.DataFrame()
test_target=pd.DataFrame()
cnt=0
while cnt<testing_data:
    j=random.randint(0,len(dataset)-1)
    new=dataset.iloc[j,:4]
    test_data=test_data.append(new)
    new_t=dataset.iloc[j,4:]
    test_target=test_target.append(new_t)
    cnt+=1
print("Number of testing data:",cnt)
print(len(test_data))
print(test_data)
print(test_target)



```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(learnset_data,learnset_target)
```

```python

print("Predictions form the classifier:")
predictions=knn.predict(test_data)

print("predicted value:")
print(predictions[:5])
print("Target value:")
print(test_target[:5])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_target,predictions)
```
